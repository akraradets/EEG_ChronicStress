```python
import os,glob,shutil
import pandas as pd
```

```python
input_path = 'data/'
output_path = 'output/'
fix_bad = True
files = glob.glob(f"{input_path}/*.csv")
if(len(files) == 0):
    raise ValueError(f"The input path you gave has no .csv files")

if(os.path.exists(output_path)):
    shutil.rmtree(output_path)
os.makedirs(output_path)
```

```python
for index, f in enumerate(files):
    print(f"===== Processing file: {index+1}/{len(files)}: {f} ====")
    df = pd.read_csv(f)
    bad_channels = []
    for channel in df.columns:
        total_drop = sum( df[channel] == -187500.016 )
        total_sample = len(df[channel])
        if(0.49 <= total_drop / total_sample <= 0.51):
            bad_channels.append(channel)
    is_bad_channels = len(bad_channels) > 0

    if(fix_bad):
        print(f"Found bad channels, attempt to fix it na ja.")
        marker = df['Marker'].copy()
        total_event_before_shift = sum(marker != '0')
        for index_inner, val in enumerate(marker):
            if(index_inner == 0): continue
            if(val != '0'): 
                print(index_inner,val)
                marker[index_inner - 1] = val
        total_event_after_shift = sum(marker != '0')
        print(total_event_before_shift, total_event_after_shift)
        df['Marker'] = marker

    else:
        raise Warning(f"Droupped signal detected in channels {bad_channels}. There are 50% of -187500.16 on the listed electrodes.\n \
            If you whish to proceed with this file, use option fix_bad=True.")

```
